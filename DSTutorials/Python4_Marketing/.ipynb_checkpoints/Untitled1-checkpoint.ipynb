```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import Google Analytics data
ga_data = pd.read_csv('data/ga_data/ga_keyword.csv')
```

```python
ga_kw_data = pd.read_csv('data/ga_data/ga_keyword.csv', skiprows=6, nrows=5)
ga_kw_traffic_data = pd.read_csv('data/ga_data/ga_keyword.csv', skiprows=14, nrows=366,
                                 parse_dates=['Day Index'])
```

```python
ga_kw_data
```

```python
ga_kw_traffic_data.head(5)
```

```python
ga_kw_traffic_data.set_index('Day Index',inplace=True)
```

```python
ga_kw_traffic_data['Users'].resample(rule='M').sum()
```

```python
#Visualizing Users by Day/Month etc
plt.plot(ga_kw_traffic_data['Users'].resample(rule='M').sum())        #By MOnth
```

```python
plt.plot(ga_kw_traffic_data['Users'].resample(rule='D').sum())      #By Day
```

```python
#Import google analytics pages data
ga_page_data = pd.read_csv('data/ga_data/ga_pages.csv', skiprows=6, nrows=376)
ga_page_traffic_data = pd.read_csv('data/ga_data/ga_pages.csv', skiprows=386, nrows=367,
                                   parse_dates=['Day Index'])
```

```python
ga_page_data.head(5)
```

```python
#VIsualizing Pages with the highest page views
ga_page_data[['Page','Pageviews']].sort_values(by='Pageviews',ascending=False).head(8)
```

```python
ga_page_data[['Page','Pageviews']].sort_values(by='Pageviews',ascending=False).head(5).plot('Page','Pageviews',kind='bar')
```
