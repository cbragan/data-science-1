```python
df['month'] = pd.to_datetime(df.loc[:,'month'])
```

```python
df.set_index('month',inplace=True)
```

```python
df.head()
```

```python
df.resample('M',kind='period')['spend'].sum()
```

```python
current_year_spend = round(df.pivot_table(index='month',values='spend',columns='format',aggfunc=sum),2)
```

```python
current_year_spend
```

```python
projections = round((current_year_spend.loc[:,'CTV']*1.25) + (current_year_spend.loc[:,'display']*1.048)+(current_year_spend.loc[:,'native']*1.048)+(current_year_spend.loc[:,'video']*1.11),2)
```

```python
projections.index=pd.date_range('2020-11-01','2021-10-01',freq='MS')
```

```python
projections
```

```python
projections.to_csv('projections.csv')
```
